This function takes in a product name as an argument, finds it's index. Then it gets a list of all similarity scores for the product index. Then it sorts the similarity scores from highest to lowest and takes only the first 30 scores and returns the product names for these indices with highest scores.

```python
def get_recommendations(Product):
    idx=indices[Product]
    sim_scores=list(enumerate(cosine_sim[idx]))
    sim_scores=sorted(sim_scores,key=lambda x:x[1],reverse=True)
    sim_scores=sim_scores[1:31]
    phone_indices=[i[0] for i in sim_scores]
    return titles.iloc[phone_indices]
```

### 10. In what business scenario you should use popularity based Recommendation Systems ?

Popolarity based system works on the principle of popularity or anything which is is in trend. These systems check about the
product or movie which are in trend or are most popular among the users and directly recommend those.Popularity based 
recommendation system works by recommending items viewed/purchased by most people and rated high. It is not personalized to particular user so, recommendations remain the same for all users. 

1.if a product is often purchased by most people then the system will get to know that the product is most popular so far every new user who just signed it,
the system will recommend that product to that user also and chances becomes high that the new user will also purchase that.

2.It helps websites to improve user engagement.

3.We can use popularity bases recommendation system when there is no data about the user and items.

4.Example: A website that stearms movies: It has listed all the movies for the users to search and watch. This results in users 
browsing through a long list of movies with no suggestions about what to watch. This, in turn, reduces the propensity of a user
to engage with the website and use its services. the simplest way to fix this issue is to use a popularity based recommendation
system. 

### 11. In what business scenario you should use CF based Recommendation Systems ?

Collaborating filtering is a method of making automatic predictions(filtering) about the interests of a user by collecting 
preferences or taste information from many users(collaborating). It filters out items that a user might like on the basis 
of reactions by similar users.

1.Its useful when we have very large data sets. We can apply collaborating filtering methods to different kinds of data
including sensing and monitoring data such as mineral exploration, environmental sensing over large areas or multiple sensors,
financial data or in E-commerce and web applications where the focus is on user data.

2.Example: If a person likes 3 movies, say Interstellar, Inception and Predestination and person B likes Inception, Predestination
and The prestige, then they have almost similar interests. We can apply CF based recommendation system in this case and come to 
a conclusion that with some certainty A should like The prestige and B should like Interstellar.
